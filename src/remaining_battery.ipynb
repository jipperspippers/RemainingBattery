```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

## Linear Model For Each Column Against Remaining Useful Life

```python
df = pd.read_csv("Featurized_Data.csv", index_col= 0)
```

```python
df
```

# Regression Based On Remaining Useful Life

```python
reg_capacity = linear_model.LinearRegression()
reg = []

for column in df:
    if(column != 'Remaining Useful Life'):
        reg.append(reg_capacity.fit(df[column].to_numpy().reshape(-1,1), df['Remaining Useful Life'].to_numpy().reshape(-1,1)))
        print( column , "Coefficients: \n", reg_capacity.coef_)
```
